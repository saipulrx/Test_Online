```python
import requests
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data API from Public API Indodax
respon_pair = requests.get('https://indodax.com/api/pairs')
respon_price_increment = requests.get('https://indodax.com/api/price_increments')
respon_summaries = requests.get('https://indodax.com/api/summaries')
respon_ticker_bitcoin = requests.get('https://indodax.com/api/ticker/btcidr')
respon_trades_bitcoin = requests.get('https://indodax.com/api/trades/btcidr')
respon_depth = requests.get('https://indodax.com/api/depth/btcidr')
```

```python
print(respon_pair.status_code)
print(respon_price_increment.status_code)
print(respon_summaries.status_code)
print(respon_ticker_bitcoin.status_code)
print(respon_trades_bitcoin.status_code)
print(respon_depth.status_code)
```

```python
#tampilkan data pair
data_pair = respon_pair.json()
df_pair = json_normalize(data_pair)
df_pair
```

```python
#tampilkan data summaries
data_summaries = respon_summaries.json()
df_summaries = json_normalize(data_summaries)
df_summaries
```

```python
df_tickers = json_normalize(data_summaries['prices_24h'])
df_tickers
```

## Menampilkan data trade

```python
df_trade = json_normalize(respon_trades_bitcoin.json())
df_trade
```

```python
df_trade['date'].unique()
```

### Konversi amount ke tipe data float

```python
amount = df_trade.amount.astype(float)
amount
```
